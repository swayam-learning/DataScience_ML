```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\hp\OneDrive\codes\python\Learn_ML\1. Datasets\cars.csv")
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder,OrdinalEncoder
```

```python
X_train,X_test,y_train,y_test = train_test_split(df.iloc[:,0:4],df.iloc[:,-1],test_size=0.2,random_state=42)
```

```python
df["owner"].value_counts()
```

```python
oe = OrdinalEncoder(categories=[["Test Drive Car","Fourth & Above Owner","Third Owner","Second Owner","First Owner"]])
X_train_owner  = oe.fit_transform(X_train.loc[:,["owner"]])
x_test_owner = oe.transform(X_train.loc[:,["owner"]])
```

```python
X_train_owner_df = pd.DataFrame(X_train_owner,columns=oe.get_feature_names_out())
x_test_owner_df = pd.DataFrame(X_test,columns=oe.get_feature_names_out())
```

```python
X_train_owner_df.head()
```

```python
ohe  = OneHotEncoder(drop="first",sparse_output=False)
```

```python
X_train_brand_fuel = ohe.fit_transform(X_train[["brand","fuel"]])
X_test_brand_fuel = ohe.transform(X_test[["brand","fuel"]])
```

```python
X_train_brand_fuel_df = pd.DataFrame(X_train_brand_fuel,columns=ohe.get_feature_names_out())
X_test_brand_fuel_df = pd.DataFrame(X_test_brand_fuel,columns=ohe.get_feature_names_out())
```

```python
X_train_brand_fuel_df.head()
```

### the better way ColummnTrabsformers_

```python
transfomer = ColumnTransformer(transformers=[
    ("trf1",OrdinalEncoder(categories=[["Test Drive Car","Fourth & Above Owner","Third Owner","Second Owner","First Owner"]]),["owner"]),
    ("trf2",OneHotEncoder(drop="first",sparse_output=False),["brand","fuel"])
],remainder="passthrough"
)
```

```python
X_train_transformed = transfomer.fit_transform(X_train)
X_test_transformed = transfomer.transform(X_test)
```

```python
pd.DataFrame(X_train_transformed,columns=transfomer.get_feature_names_out())
```
