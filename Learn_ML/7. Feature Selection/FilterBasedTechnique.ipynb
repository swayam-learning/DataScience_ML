#### Duplicate Features

````python
def get_duplicate_columns(df):
    
    duplicate_columns = {}
    seen_columns = {}

    for column in df.columns:
        current_column = df[column]

        # Convert column data to bytes
        try:
            current_column_hash = current_column.values.tobytes()
        except AttributeError:
            current_column_hash = current_column.to_string().encode()

        if current_column_hash in seen_columns:
            if seen_columns[current_column_hash] in duplicate_columns:
                duplicate_columns[seen_columns[current_column_hash]].append(column)
            else:
                duplicate_columns[seen_columns[current_column_hash]] = [column]
        else:
            seen_columns[current_column_hash] = column

    return duplicate_columns
````

````python
duplicate_columns = get_duplicate_columns(X_train)
````


<h3>Variance Threshold Method</h3>

<li>Constant Feature
<p>in constant method , for eg: we have 3 columns A,B,C the column B thorughout has same value so it does not make any sense to use it in predicting C so we drop it</p></li>

<li>Quasi Constant Feature
<p>
In Quasi Constant method , for eg: we have 1000 rows and the column B has values constant for, like 950 rows so the variance in be is nearly zero again so we can drop it as it does not add any value in prediciting C
</p></li>


````python
from sklearn.feature_selection import VarianceThreshold
sel = VarianceThreshold(threshold=0.05)
````

````python
sel.fit(X_train)
````

````python
sel.get_support()
# it is used to check which columns have variance less than the threshold we provided her 0.05
````

````python
sum(sel.get_support())
````

````python
columns = X_train.columns[sel.get_support()]
````

````python
columns
````

````python
# now we are deleting the columns which have less than the threshold
X_train = sel.transform(X_train)
X_test = sel.transform(X_test)

X_train = pd.DataFrame(X_train,columns=columns)
X_test = pd.DataFrame(X_test,columns=columns)
````

````python
print(X_train.shape)
print(X_test.shape)
````

<h3>Correlation</h3>
<li> -1 ---> Strong Inverese Linear relationship </li>
<li>  1 ----> Strong Positive Linear relationship </li>

````python
import seaborn as sns
sns.heatmap(X_train.corr())
````

````python
corr_matrix  = X_train.corr()
````

````python
columns  = corr_matrix.columns
columns_to_drop = []
for i in range(len(columns)):
    for j in range(i+1,len(columns)):
        if corr_matrix.loc[columns[i],columns[j]] > 0.95:
            columns_to_drop.append(columns[j])
print(len(columns_to_drop))
````

````python
columns_to_drop = set(columns_to_drop)
````

````python
len(columns_to_drop)
````

````python
X_train.drop(columns= columns_to_drop,axis=1,inplace=True)
X_test.drop(columns= columns_to_drop,axis=1,inplace=True)
print(X_train.shape)
print(X_test.shape)
````
