#### Gradient Descent
B new = B old - n*slope


##### n * slope is called as step size, n being the learning rate

```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
X,y = make_regression(n_samples=4 ,n_features=1,n_informative=1,n_targets=1,noise=80,random_state=13)
# we have n_features = 1 that is 1 column
# n_samples = 4 means ,4 rows
# n_informative = 1 ,1 feature affects in the making of target variable
# noise = 80 the amt of std used to make target variable 
```

```python
X
```

```python
y
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(X,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color="red")
plt.show()
```

### Iteration 0

```python
# for a slope of 78.35 now we try with intercept b = 0\
b = 0
y_pred = (78.35*X+b).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color="red",label="Ordinary Least Square")
plt.plot(X,y_pred,color="blue",label="b=0")
plt.legend()
plt.show()
```

#### fromula for slope


### Iteration 1

#### Calculating Loss slope

```python
m=78.35
loss_slope = -2 * np.sum(y-m*X.ravel()-b)
loss_slope
```

#### Calculating the step size

```python
# learning rate
lr = 0.1 

step_size = loss_slope*lr
step_size
```

#### Calculating the new intercept

```python
b = b - step_size
# b new = b old - loss_slope * learning_rate 
b
```

```python
y_pred1 = ((78.35*X)+b).reshape(4)
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color="red",label="Ordinary Least Square")
plt.plot(X,y_pred1,color="#00a65a",label="b={}".format(b))
plt.plot(X,y_pred,color="#A3E4D7",label="b=0")
plt.legend()
plt.show()
```

### Iteration 2

### Repeating the same steps again for better results

```python
loss_slope = -2 * np.sum(y-m*X.ravel()-b)
loss_slope
```

```python
lr = 0.1 

step_size = loss_slope*lr
step_size
```

```python
b = b - step_size
# b new = b old - loss_slope * learning_rate 
b
```

```python
y_pred2 = ((78.35*X)+b).reshape(4)
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color="red",label="Ordinary Least Square")
plt.plot(X,y_pred2,color="#00a65a",label="b={}".format(b))
plt.plot(X,y_pred1,color="#A3E4D7",label="b=20.927")
plt.plot(X,y_pred,color="#A3E4D7",label="b=0")
plt.legend()
plt.show()
```
