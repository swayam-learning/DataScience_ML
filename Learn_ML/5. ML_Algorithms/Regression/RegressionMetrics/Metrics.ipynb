```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv(r"C:\Users\hp\OneDrive\codes\python\Learn_ML\1. Datasets\placement.csv")
df.head()
```

```python
X_train,X_test,y_train,y_test = train_test_split(df.iloc[:,0:1],df.iloc[:,1],test_size=0.2,random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
plt.scatter(df["cgpa"],df["package"])
plt.plot(X_train,lr.predict(X_train),c="red")
plt.xlabel("CGPA")
plt.ylabel("Package")
plt.show()
```

```python
y_pred = lr.predict(X_test)
y_pred
```

```python
print("R2 score",r2_score(y_test,y_pred))
print("Mean absolute error",mean_absolute_error(y_test,y_pred))
print("Mean squared error",mean_squared_error(y_test,y_pred))
```

```python
r2 = r2_score(y_test,y_pred)
r2
```

## Adjusted r2 score

```python
1-((1-r2)*(40-1)/((40-1-1)))
```

#### there is not a significant difference between r2 score and adjusted r2 score
