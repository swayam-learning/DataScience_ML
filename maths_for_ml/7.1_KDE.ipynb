## KDE from sklearn.neighbours

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
from numpy.random import normal
sample1 = normal(loc=20,scale=5,size=300)
sample2 = normal(loc=50,scale=5,size=700)
sample = np.hstack((sample1,sample2))
```

```python
sample
```

```python
from sklearn.neighbors import KernelDensity
model = KernelDensity(bandwidth = 1,kernel = "gaussian")
# bandwidth the std of the smaller curves of the x points of the kde plot
sample = sample.reshape((len(sample),1))
model.fit(sample)
```

```python
values = np.linspace(sample.min(),sample.max(),100)
values = values.reshape(len(values),1)
```

```python
probaility_density = model.score_samples(values)
probaility_density = np.exp(probaility_density)
```

##### the score_samples() gives the answer as log so to reverse it we use np.exp() to get actual probaility density values

```python
plt.hist(sample,bins=50,density=True)
plt.plot(values,probaility_density)
plt.show()
```

```python
sns.kdeplot(sample,bw_adjust=2)
# bw_adjust is same as bandwidth
plt.show()
```
