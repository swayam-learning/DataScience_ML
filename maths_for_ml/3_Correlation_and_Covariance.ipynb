```python
import numpy as np
import matplotlib.pyplot as plt

# Simulate rolling a six-sided dice 100 times
dice_rolls = np.random.randint(1, 7, size=1000)

# Sort the outcomes
sorted_outcomes = np.sort(dice_rolls)

# Compute the cumulative probabilities
cum_prob = np.linspace(0, 1, len(sorted_outcomes))

# Plot the CDF
plt.figure(figsize=(8, 6))
plt.step(sorted_outcomes, cum_prob, where='post')
plt.title('Cumulative Density Function (CDF) of Dice Rolls')
plt.xlabel('Dice Outcomes')
plt.ylabel('Cumulative Probability')
plt.xticks(range(1, 7))
plt.yticks(np.linspace(0, 1, 11))
plt.grid(True)
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a sample dataset (normal distribution for example)
data = np.random.normal(loc=0, scale=1, size=1000)
# loc == mean 
# scale == standard deviation
# Sort the data
sorted_data = np.sort(data)

# Compute the cumulative probabilities
cum_prob = np.linspace(0, 1, len(sorted_data))

# Plot the CDF
plt.figure(figsize=(8, 6))
plt.plot(sorted_data, cum_prob, marker='.', linestyle='none')
plt.title('Cumulative Density Function (CDF)')
plt.xlabel('Data')
plt.ylabel('Cumulative Probability')
plt.grid(True)
plt.show()
```
