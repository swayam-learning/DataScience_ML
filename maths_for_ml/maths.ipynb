```python
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
```

### Two die thrown 10000 times and their sum ranges from 2 to 12, theoretically the probability for 2 and 12 should be least and probability for 7 should be highest

```python
l=[]
for i in range(10000):
    a = random.randint(1,6)
    b= random.randint(1,6)
    l.append(a+b)
```

```python
l[:5]
```

```python
s = (pd.Series(l).value_counts()/pd.Series(l).value_counts().sum()).sort_index()
s
```

### Probability Mass Function

```python
s.plot(kind="bar")
```

### Cummulative Mass Function

```python
np.cumsum(s)
```

```python
np.cumsum(s).plot(kind="bar")
```

### Probability Density Function

```python
import seaborn as sns
sns.kdeplot(s)
```
