```python
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
```

### Two die thrown 10000 times and their sum ranges from 2 to 12, theoretically the probability for 2 and 12 should be least and probability for 7 should be highest

```python
l=[]
for i in range(10000):
    a = random.randint(1,6)
    b= random.randint(1,6)
    l.append(a+b)
```

```python
l[:5]
```

```python
s = (pd.Series(l).value_counts()/pd.Series(l).value_counts().sum()).sort_index()
s
```

### Probability Mass Function

```python
s.plot(kind="bar")
```

### Cummulative Mass Function

```python
np.cumsum(s)
```

```python
np.cumsum(s).plot(kind="bar")
```

### Probability Density Function

```python
import seaborn as sns
sns.kdeplot(s)
```

### Cummulative Density Function

```python
a= np.cumsum(s)
a.plot()
```

### Paramteric Density Function

```python
from numpy.random import normal
sample  = normal(loc=50,scale =5 , size =1000)
# loc = mean = 50 
# scale = std = 5
```

```python
sample.mean() # nearly close to 50
```

```python
sample.std() # nearly close to 5
```

```python
# histogram
plt.hist(sample)
plt.show()
```

```python
sample_mean = sample.mean()
sample_std = sample.std()
```

```python
# fitting the mean , std on normal distribution function
from scipy.stats import norm
dist = norm(sample_mean,sample_std)
```

```python
values  = np.linspace(sample.min(),sample.max(),100) # we generated 100 x values (the support values), from minimumm to maximum
```

```python
probability_density = [dist.pdf(value) for value in values]
```

```python
plt.hist(sample,bins=10,density=True)
plt.plot(values,probability_density)
plt.show()
```

This whole process is called paramteric density function because it depends on the two parameters like mean and standard deviation
