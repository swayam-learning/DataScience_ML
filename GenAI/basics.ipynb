### What is LLM(Large Language Model) ?

LLM's are foundational machine learning models that use deep learning algorithms to process and understand natural language .The models are trained on massive amounts of text data to learn patterns and entity relationships in the language.

It is a language model which is responsible for performing task such as text to text generation , text to image generation and image to text generations.
