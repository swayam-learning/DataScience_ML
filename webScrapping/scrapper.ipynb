```python
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import requests
```

```python
url = "https://quotes.toscrape.com/"
html = requests.get(url)
html
```

```python
soup  = BeautifulSoup (html.text,"html.parser",)
print(soup.prettify())
```

```python
soup.body.prettify
```

```python
title = soup.title
title
```

## Finding the Quotes from the page

```python
all_title = soup.find_all("span",class_="text")
all_title
```

```python
for quotes in all_title:
    print(quotes.text[1:-1])
```

## Finding all the author names

```python
authors = soup.find_all("small",class_="author")
for author in authors:
    print(author.text[0:])
```

## Finding the tags from all the quotes

```python
for i in soup.find_all("div",class_="quote"):
    tags = []
    for tag in i.find_all("a",class_="tag"):
        tags.append(tag.text)
    tags = ",".join(tags)
    print(tags)
```

```python
data1 = []
for i in soup.find_all("div",class_="quote"):
    quotes = i.find("span",class_="text").text[1:-1]
    authors = i.find("small",class_="author").text[0:]
    tags = []
    for tag in i.find_all("a",class_="tag"):
        tags.append(tag.text)
    tags = ",".join(tags)
    data1.append([quotes,authors,tags])
```

```python
data1
```

```python
df = pd.DataFrame(data1,columns=["Quotes","Authors","Tags"])
df
```

```python
df.to_html("Your Html For Quotes.html",justify="left")
```

```python
df.to_csv("Quotes.csv")
```
