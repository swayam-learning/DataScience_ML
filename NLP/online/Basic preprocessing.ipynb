### 5. Stemming and lamitization

```python
from nltk.stem import LancasterStemmer, WordNetLemmatizer
```

#### 5.1 Stemming

```python
from nltk.stem import LancasterStemmer
import re

#Initialize the Lancaster Stemmer
lancaster = LancasterStemmer()

#Given text
text = "It took them the best part of three years to work out how to do it. Your father and Sirius here were the cleverest students in the school, and lucky they were, because the Animagus transformation can go horribly wrong."

#Tokenizing words(removing punctuation and splitting)
words = re.findall(r'\b\w+\b', text)

#Applying Lancaster Stemming
stemmed_words = [lancaster.stem(i) for i in words]

#Display the results
print("Original Words: ", words)
print("Stemmed Words: ", stemmed_words)
```

####    5.2 Lemmatization

```python
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
import nltk

#ownload wordnet data if not already downloaded
nltk.download('wordnet')
nltk.download('punkt')

#Initialize WordNet Lemmatizer
lemmatizer = WordNetLemmatizer()

#Given sentence
sentence = "It took them the best part of three years to work out how to do it. Your father and Sirius here were the cleverest students in the school, and lucky they were, because the Animagus transformation can go horribly wrong."

#Tokenizing words
words = word_tokenize(sentence.lower()) #convert to lowercase for consistency

#Applying lemmmatization(default POS = noun, better results with verb tagging)
lemmatized_words = [lemmatizer.lemmatize(word, pos="v") for word in words]

#Displaying results
print("Original Sentence: ", sentence)
print("Lemmatized Sentence: ", " ".join(lemmatized_words))

```
