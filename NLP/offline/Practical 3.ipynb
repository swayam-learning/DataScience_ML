### Stop words
stop words are words of little value in text analysis because they occur frequently in the language and typically do not carry meaning on their own

```python
import nltk
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

#download nltk data
nltk.download('punkt')
nltk.download('stopwords')
                            ###Punkt and stopwords are pre-trained datasets

sentence = "Stop words are common words that are often considered to be of little value in text analysis."

#tokenize then sentence
words = word_tokenize(sentence)

#get the list of english stop words from NLTK
stop_words = set(stopwords.words('english'))

#Remove stop words from the tokenized words
filtered_words = [word for word in words if word.lower() not in stop_words]

#display the original and filtered words
print("Original Words: ", words)
print("Filtered Words: ", filtered_words)
```

*Ques: "The sun was shining brightly in the sky, and a gentle breeze was blowing through the trees". Write a python program to remove all stop words(e.g., "a", "the", "was", "and") from the text and return the cleaned version*

```python
#Answer :
sentence = "The sun was shining brightly in the sky, and a gentle breeze was blowing through the trees"

words = word_tokenize(sentence)

stop_words = set(stopwords.words('english'))

filtered_words = [word for word in words if word.lower() not in stop_words]

print("Original Words: ", words)
print("Filtered Words: ", filtered_words)
```

```python
import spacy
```

```python
#load spacy English model
nlp = spacy.load("en_core_web_sm")
```

```python
#Add Custom stopwords
nlp.Defaults.stop_words.add("nil")
nlp.Defaults.stop_words.add("junk")
```

```python
#Sample text
text = "This is a JUNK sentence that contains NIL information but is useful for texting"
```

```python
#process text with spacy
doc = nlp(text)
```

```python
#remove stopwords
filtered_words = [token.text for token in doc if token.text.lower() not in nlp.Defaults.stop_words]
```

```python
#Display results
print("Original text: ", text)
print("Filtered text: "," ".join(filtered_words))
```
