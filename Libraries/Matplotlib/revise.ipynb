## Line Plot

### Track Record of Virat Kohli and R Sharma in ipl

```python
plt.plot(batsman['index'],batsman['V Kohli'],color="green")
plt.plot(batsman['index'],batsman['RG Sharma'],color="black")
plt.title("Rohit Sharma vs Virat kohli")
plt.ylabel("Runs Scored")
plt.xlabel("Year")
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color="green",linestyle="dashdot",linewidth=3,marker="D",markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],color="black",linestyle="dashdot",linewidth=2)
plt.title("Rohit Sharma vs Virat kohli")
plt.ylabel("Runs Scored")
plt.xlabel("Year")
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color="green",linestyle="dashdot",linewidth=3,marker="D",markersize=10,label="Virat")
plt.plot(batsman['index'],batsman['RG Sharma'],color="black",linestyle="dashdot",linewidth=2,label="Rohit")
plt.title("Rohit Sharma vs Virat kohli")
plt.ylabel("Runs Scored")
plt.xlabel("Year")
plt.legend(loc="best")
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color="green",linestyle="dashdot",linewidth=3,marker="D",markersize=10,label="Virat")
plt.plot(batsman['index'],batsman['RG Sharma'],color="black",linestyle="dashdot",linewidth=2,label="Rohit")
plt.title("Rohit Sharma vs Virat kohli")
plt.ylabel("Runs Scored")
plt.xlabel("Year")
plt.legend(loc="best")
plt.grid()
plt.show()
```
