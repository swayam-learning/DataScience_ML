```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x= np.linspace(0,10,200)
```

```python
x
```

```python
y = np.sin(x)
```

```python
y
```

### Plot 

```python
plt.figure(figsize=(20.4,5.4))
plt.plot(x,y,'--g')
plt.xlabel("this is x axis")
plt.ylabel("this is y aixs")
plt.title("Plotting x v/s y")
plt.grid()
plt.show()
```

### Scatter Plot

```python
x = np.random.rand(50)
y = np.random.randn(50)
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y,c="g",s=x*100,alpha=0.6,marker="x") # c for color , s for sizes and alpha for opacity , marker for our own marker
plt.xlabel("this is x axis")
plt.ylabel("this is y aixs")
plt.title("Plotting x v/s y")
plt.show()
```

```python
plt.scatter(x,y,c=x,s=x*100,alpha=1.0) # c for color , s for sizes and alpha for opacity
plt.xlabel("this is x axis")
plt.ylabel("this is y aixs")
plt.title("Plotting x v/s y")
plt.show()
```

### Bargraph 
#### categorical vs numerical

```python
x = ['a','c','d','f','g']
y=np.random.rand(5)
```

```python
y
```

```python
plt.bar(x,y)
plt.xlabel("x axis ---->")
plt.ylabel("y axis ---->")
```

```python
plt.barh(x,y) # barh for horizontal bar plot
plt.xlabel("x axis ---->")
plt.ylabel("y axis ---->")
```
