```python
df = pd.read_csv(r"C:\Users\hp\OneDrive\codes\python\Learn_ML\1. Datasets\covid_toy.csv")
df.head()
```

```python
X_train,X_test,y_train,y_test = train_test_split(df.iloc[:,0:5],df.iloc[:,-1],test_size=0.8,random_state=42)
```

```python
transformers = ColumnTransformer(transformers=[
                                 ('trf1',SimpleImputer(strategy='most_frequent'),['fever']),
                                 ('trf2',OneHotEncoder(drop="first",sparse_output=False),['gender','city']),
                                 ('trf3',OrdinalEncoder(categories=[['Mild','Strong']]),['cough'])],remainder='passthrough')
```

```python
transformers.fit_transform(X_train)
```
