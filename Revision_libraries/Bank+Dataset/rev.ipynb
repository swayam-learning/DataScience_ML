```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load sample dataset (tips dataset)
df = sns.load_dataset('tips')

### UNIVARIATE ANALYSIS ###

# 1. Histogram (Univariate for 'total_bill')
plt.figure(figsize=(8, 6))
plt.hist(df['total_bill'], bins=10, color='skyblue')
plt.title('Histogram of Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.show()

# 2. Boxplot (Univariate for 'total_bill')
plt.figure(figsize=(8, 6))
sns.boxplot(df['total_bill'])
plt.title('Boxplot of Total Bill')
plt.show()

# 3. Barplot (Univariate for 'day' count)
plt.figure(figsize=(8, 6))
sns.countplot(x='day', data=df)
plt.title('Barplot of Days')
plt.show()

# 4. Pie Chart (Univariate for 'smoker' category)
plt.figure(figsize=(8, 6))
df['smoker'].value_counts().plot.pie(autopct='%1.1f%%', colors=['skyblue', 'lightcoral'])
plt.title('Pie Chart of Smoker Category')
plt.ylabel('')  # Remove the y-label
plt.show()


### BIVARIATE ANALYSIS ###

# 5. Boxplot (Bivariate - Total Bill by Day)
plt.figure(figsize=(8, 6))
sns.boxplot(x='day', y='total_bill', data=df)
plt.title('Boxplot of Total Bill by Day')
plt.show()

# 6. Violin Plot (Bivariate - Total Bill by Time)
plt.figure(figsize=(8, 6))
sns.violinplot(x='time', y='total_bill', data=df)
plt.title('Violin Plot of Total Bill by Time')
plt.show()

# 7. Scatter Plot (Bivariate - Total Bill vs Tip)
plt.figure(figsize=(8, 6))
sns.scatterplot(x='total_bill', y='tip', data=df)
plt.title('Scatter Plot of Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()

# 8. Pair Plot (Bivariate Relationships for Numeric Columns)
sns.pairplot(df)
plt.suptitle('Pair Plot of Numeric Variables', y=1.02)
plt.show()

# 9. Heatmap (Bivariate Correlation Matrix for Numeric Columns)
numeric_df = df.select_dtypes(include=['float64', 'int64'])
corr_matrix = numeric_df.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Heatmap of Correlation Matrix (Numeric Variables Only)')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load sample dataset (tips dataset)
df = sns.load_dataset('tips')

# Regplot for multivariate analysis: Total Bill vs Tip with Smoker as hue
plt.figure(figsize=(8, 6))
sns.regplot(x='total_bill', y='tip', data=df, scatter_kws={'color':'skyblue'}, line_kws={'color':'black'})
plt.title('Regplot of Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()

# Another way: Introducing a third variable with 'hue' in a lmplot (for categorical data)
sns.lmplot(x='total_bill', y='tip', hue='smoker', data=df, aspect=1.5)
plt.title('Regplot with Hue (Smoker vs Non-Smoker)')
plt.show()
```

```python
df = pd.read_csv("bank_marketing_updated_v1.csv",skiprows=2)
df.head()
```
