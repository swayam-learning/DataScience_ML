## Broadcasting 
#### -- done by numpy automatically here we are trying to understand where we can do this

##### The term broadcasting how numpy treats arrays with different shapes during arithemtic operations
##### the smaller array is broadcast across the larger array so that they have compatible shapes

```python
a = np.arange(6).reshape(2,3)
b = np.arange(3).reshape(1,3)

print(a,"\n")
print(b,"\n")

print(a*b)
```

#### Broadcasting Rules

**1. Make the two arrays have the same number of dimensions.**<br>
- If the numbers of dimensions of the two arrays are different, add new dimensions with size 1 to the head of the array with the smaller dimension.<br>



**2. Make each dimension of the two arrays the same size.**<br>
- If the sizes of each dimension of the two arrays do not match, dimensions with size 1 are stretched to the size of the other array.
- If there is a dimension whose size is not 1 in either of the two arrays, it cannot be broadcasted, and an error is raised.



```python
import numpy as np
```

```python
np.ones((3,3)) + np.arange(2)
```

```python
a = np.arange(0,12).reshape(4,3)
b = np.arange(3)

print(a,"\n")
print(b,"\n")

print(a+b)
```

```python
a = np.arange(0,12).reshape(3,4)
b = np.arange(3)

print(a,"\n")
print(b,"\n")

print(a+b)
```

```python
a = np.arange(3).reshape(3,1)
b = np.arange(3).reshape(1,3)

print(a,"\n")
print(b,"\n")

print(a+b)
```

```python
a = np.array([1])
# shape changed to 2,2
b = np.arange(4).reshape(2,2)

print(a,"\n")
print(b,"\n")

print(a+b)
```

##### broadcasting errors for more clarity on these concepts

```python
a = np.arange(12).reshape(3,4)
b = np.arange(12).reshape(4,3)

print(a)
print(b)
# shape cannot be changed as we dont have any 1 in either of the arrays
print(a+b)
```

```python
a = np.arange(16).reshape(4,4)
b = np.arange(4).reshape(2,2)

print(a)
print(b)
# shape cannot be changed as we dont have any 1 in either of the arrays
print(a+b)
```

##### Mean Squared Error used as loss function for linear regression



```python
actual = np.random.randint(1,50,25)
predicted = np.random.randint(1,50,25)
```

```python
actual
```

```python
predicted
```

##### just assume this was a dataset where we expected some result and the predicted result result was diiferent so now in ordee to deal with this loss we use mean squared error

```python
def msc(y,y1):
    
    return np.mean((y-y1)**2)

y = actual
y1 = predicted

msc(y,y1)
```

#### Binary Cross Entropy




```python
def bce(y_actual,y_predicted):
    epsilon = 1e-15  # small value to prevent log(0)
    y_predicted = np.clip(y_predicted, epsilon, 1 - epsilon) # clip to prevent log(0)
    return -np.mean(y_actual*np.log(y_predicted)+(1-y_actual)*np.log(1-y_predicted))

y_actual = actual
y_predicted = predicted

bce(y_actual,y_predicted)
```
